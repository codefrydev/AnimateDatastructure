```
using System;
using System.Collections.Generic;
using System.Text;
using System.Linq;
using System.IO;
```

```
public enum FolderName
{
    Greedy,
    Matrix,
    SlidingWindow,
    Tree,
    Array,
    Stack,
    DP,
    Linear,
    BinarySearch,
    TwoPointer
}
```

```
var folder = FolderName.TwoPointer;

// Edit these — LeetCode URL slug is the path segment on leetcode.com/problems/<slug>/
var problemLeetCodeUrlSlug = "minimum-absolute-distance-between-mirror-pairs";
var problemDir = Path.Combine(Directory.GetCurrentDirectory(), folder.ToString(), problemLeetCodeUrlSlug);
// File basename for .html / .json (use problemLeetCodeUrlSlug, or e.g. $"{problemLeetCodeUrlSlug}-binary" for another viz in the same folder)
var problemFileName = problemLeetCodeUrlSlug;
var problemName = "Minimum Absolute Distance Between Mirror Pairs";

problemDir
```

```
public class ProblemData
{
    public string ProblemUri { get; set; }
    public string ProblemName { get; set; }
    public string ProblemDescription { get; set; }
    public string ProblemSolution { get; set; }
    public string Difficulty { get; set; }
    public List<string> Topics { get; set; }
    public int ProblemNumber { get; set; }
}
```

```
Directory.CreateDirectory(problemDir);

var jsonPath = Path.Combine(problemDir, problemFileName + ".json");
var htmlPath = Path.Combine(problemDir, problemFileName + ".html");

if (!File.Exists(jsonPath))
{
    var data = new ProblemData
    {
        ProblemUri = $"https://leetcode.com/problems/{problemLeetCodeUrlSlug}",
        ProblemName = problemName,
        ProblemDescription = """


You are given two non-increasing 0-indexed integer arrays nums1​​​​​​ and nums2​​​​​​.

A pair of indices (i, j), where 0 <= i < nums1.length and 0 <= j < nums2.length, is valid if both i <= j and nums1[i] <= nums2[j]. The distance of the pair is j - i​​​​.

Return the maximum distance of any valid pair (i, j). If there are no valid pairs, return 0.

An array arr is non-increasing if arr[i-1] >= arr[i] for every 1 <= i < arr.length.

 

 
""",
        ProblemSolution = folder.ToString(),
        Difficulty = "Medium",
        Topics = """


Senior
Array
Two Pointers
Binary Search
Weekly Contest 240

""".Split([' ','\n']).Select(x=>x.Trim()).Where(x=>x.Length>0).ToList(),


        ProblemNumber = 1855
    };

    var json = System.Text.Json.JsonSerializer.Serialize(data, new System.Text.Json.JsonSerializerOptions
    {
        WriteIndented = true
    });

    File.WriteAllText(jsonPath, json);
}

if (!File.Exists(htmlPath))
{
    File.WriteAllText(htmlPath, "");
}
```
