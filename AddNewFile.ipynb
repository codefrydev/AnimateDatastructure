```
using System;
using System.Collections.Generic;
using System.Text;
using System.Linq;
using System.IO;
```

```
public enum FolderName
{
    Greedy,
    Matrix,
    SlidingWindow,
    Tree,
    Array,
    Stack,
    DP,
    Linear
}
```

```
var folder = FolderName.Linear;
folder
```

```
var fileName = "minimum-distance-to-the-target-element";
var folderName = folder.ToString();
folderName
```

```
var dir = Directory.GetCurrentDirectory();
```

```
var newFileAndFolderPath = Path.Combine(dir , folderName);
newFileAndFolderPath
```

```
public class ProblemData
{
    public string ProblemUri { get; set; }
    public string ProblemName { get; set; }
    public string ProblemDescription { get; set; }
    public string ProblemSolution { get; set; }
    public string Difficulty { get; set; }
    public List<string> Topics { get; set; }
    public int ProblemNumber { get; set; }
}
```

```
Directory.CreateDirectory(newFileAndFolderPath);

var jsonPath = Path.Combine(newFileAndFolderPath, fileName + ".json");
var htmlPath = Path.Combine(newFileAndFolderPath, fileName + ".html");

if (!File.Exists(jsonPath))
{
    var data = new ProblemData
    {
        ProblemUri = $"https://leetcode.com/problems/{fileName}",
        ProblemName = "Minimum Distance to the Target Element",
        ProblemDescription = """
Given an integer array nums (0-indexed) and two integers target and start, find an index i such that nums[i] == target and abs(i - start) is minimized. Note that abs(x) is the absolute value of x.

Return abs(i - start).

It is guaranteed that target exists in nums.
""",
        ProblemSolution = folderName,
        Difficulty = "Easy",
        Topics = """
Mid Level
Array
Weekly Contest 239
""".Split([' ','\n']).Select(x=>x.Trim()).Where(x=>x.Length>0).ToList(),


        ProblemNumber = 1848
    };

    var json = System.Text.Json.JsonSerializer.Serialize(data, new System.Text.Json.JsonSerializerOptions
    {
        WriteIndented = true
    });

    File.WriteAllText(jsonPath, json);
}

if (!File.Exists(htmlPath))
{
    File.WriteAllText(htmlPath, "");
}
```
