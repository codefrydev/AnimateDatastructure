```
using System;
using System.Collections.Generic;
using System.Text;
using System.Linq;
using System.IO;
```

```
var fileName = "lexicographically-smallest-generated-string";
var folderName = "Greedy";
```

```
var dir = Directory.GetCurrentDirectory();
```

```
var newFileAndFolderPath = Path.Combine(dir , folderName);
newFileAndFolderPath
```

```
public class ProblemData
{
    public string ProblemUri { get; set; }
    public string ProblemName { get; set; }
    public string ProblemDescription { get; set; }
    public string ProblemSolution { get; set; }
    public string Difficulty { get; set; }
    public List<string> Topics { get; set; }
    public int ProblemNumber { get; set; }
}
```

```
Directory.CreateDirectory(newFileAndFolderPath);

var jsonPath = Path.Combine(newFileAndFolderPath, fileName + ".json");
var htmlPath = Path.Combine(newFileAndFolderPath, fileName + ".html");

if (!File.Exists(jsonPath))
{
    var data = new ProblemData
    {
        ProblemUri = $"https://leetcode.com/problems/{fileName}",
        ProblemName = "Lexicographically Smallest Generated String",
        ProblemDescription = """

You are given two strings, str1 and str2, of lengths n and m, respectively.

A string word of length n + m - 1 is defined to be generated by str1 and str2 if it satisfies the following conditions for each index 0 <= i <= n - 1:

    If str1[i] == 'T', the substring of word with size m starting at index i is equal to str2, i.e., word[i..(i + m - 1)] == str2.
    If str1[i] == 'F', the substring of word with size m starting at index i is not equal to str2, i.e., word[i..(i + m - 1)] != str2.

Return the Lexicographically Smaller possible string that can be generated by str1 and str2. If no string can be generated, return an empty string "".


A substring is a contiguous non-empty sequence of characters within a string.

A string a is lexicographically smaller than a string b if in the first position where a and b differ, string a has a letter that appears earlier in the alphabet than the corresponding letter in b.
If the first min(a.length, b.length) characters do not differ, then the shorter string is the lexicographically smaller one.
""",
        ProblemSolution = "Greedy String Matching",
        Difficulty = "Hard",
        Topics = """

Principal
String
Greedy
String Matching
Weekly Contest 439

""".Split([' ','\n']).Select(x=>x.Trim()).Where(x=>x.Length>0).ToList(),


        ProblemNumber = 3474
    };

    var json = System.Text.Json.JsonSerializer.Serialize(data, new System.Text.Json.JsonSerializerOptions
    {
        WriteIndented = true
    });

    File.WriteAllText(jsonPath, json);
}

if (!File.Exists(htmlPath))
{
    File.WriteAllText(htmlPath, "");
}
```
